### 9. Rela√ß√£o quantidade de usu√°rios por quantidade de mensagens por estado;

```python
conn2.execute(
'''
SELECT 
    state, 
    COUNT(text_content_anonymous) / COUNT(distinct id_member_anonymous) as relacao_usuario_mensagem
FROM tabela_wpp 
GROUP BY state
'''
).fetchdf()
```

### 10. Quantidade de mensagens por pa√≠s;

```python
conn2.execute(
'''
SELECT 
    country, 
    COUNT(text_content_anonymous) as qnt_mensagem
FROM tabela_wpp 
GROUP BY country
'''
).fetchdf()
```

### 11. Quantidade de mensagens Brasil X Pa√≠ses Estrangeiros;

```python
conn2.execute(
'''
SELECT 
    SUM(case when country = 'BRAZIL' then 1 else 0 end) as qnt_brasil,
    SUM(case when country = 'BRAZIL' then 0 else 1 end) as qnt_estrangeiros
FROM tabela_wpp 
WHERE text_content_anonymous is not null
'''
).fetchdf()
```

### 21. Os 30 unigramas, bigramas e trigramas mais compartilhados;

```python
conn2.execute(
'''
SELECT 
    text_content_anonymous,
    count(*) as contagem
FROM tabela_wpp 
WHERE has_media = false and has_media_url = false
GROUP BY text_content_anonymous
HAVING array_length(string_split(lower(text_content_anonymous), ' ')) <= 3
ORDER BY contagem DESC
'''
).fetchdf()
```

### 22. As 30 mensagens mais positivas (distintas);

```python
conn2.execute(
'''
SELECT 
    distinct text_content_anonymous,
    avg(score_sentiment) as avaliacao
FROM tabela_wpp 
GROUP BY text_content_anonymous
HAVING text_content_anonymous is not null
ORDER BY avaliacao DESC
fetch next 30 row only
'''
).fetchdf()
```

### 28. O dia em que foi publicado a maior quantidade de mensagens;

```python
conn2.execute(
'''
SELECT 
    DATE_TRUNC('day', date_message), 
    COUNT(*) as quantidade
FROM tabela_wpp 
GROUP BY DATE_TRUNC('day', date_message)
ORDER BY quantidade DESC
fetch next 1 row only
'''
).fetchdf()
```
