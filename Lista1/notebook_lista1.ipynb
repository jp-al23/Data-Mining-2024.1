## Q1.k

```python
def verificar_sentimento(linha):
    if linha['midia'] == 1 or linha['url'] == 1: return 0
    elif type(linha['text']) != str: return 0
    elif linha['words'] > 30 or linha['text'].count('\n') > 3: return 0
    elif any(substring in linha['text'].lower() for substring in ["oi", "suave", "eai", "como vai", "bom dia", ":)", ":3", "olá", "kk", "ksks", "oie", 'eae', 'ótimo', 'otimo', 'bem']): return 1
    elif any(substring in linha['text'].lower() for substring in ["morreu", ":(", "triste", 'sad', ':/', ';-;', 'mal']): return -1
    else: return 0

df['sentiment'] = df.apply(verificar_sentimento, axis = 1)
df
```

```python
def verificar_travazap(linha):
    if type(linha['text']) != str: return False
    elif linha['words'] > 30 or linha['text'].count('\n') > 3: return True
    else: return False

df_filtrado = df[ ~df.apply(verificar_travazap, axis = 1)]
exemplos_travazap = df[ df.apply(verificar_travazap, axis = 1)]['text']
df_filtrado
```
